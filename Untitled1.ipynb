```python
  import pandas as pd
  import matplotlib.pyplot as plt
  import seaborn as sns
  df =  pd.read_csv('student-mat.csv', sep=r'\s*;\s*',
                           header=0, encoding='ascii', engine='python')
  print("Col:", df.columns)
  df.columns = df.columns.str.strip()
  print(df.head())
  miss = df.isnull().sum()
  print(miss)
  datatypoe = df.dtypes
  print(datatypoe)
  shape = df.shape
  print(shape)
  df.dropna(inplace=True)
  df.drop_duplicates(inplace=True)
  average = df['G3'].mean()
  print(f'Average math score: {average}')
  studentsabv = df[df['G3'] > 15].shape[0]
  print(f'above 15: {studentsabv}')
  corren = df['studytime'].corr(df['G3'])
  print(f'study time and final grade: {corren}')
  gradegender = df.groupby('sex')['G3'].mean()
  print("Average by gender:\n", gradegender)
  plt.figure(figsize=(10, 6))
  plt.hist(df['G3'], bins=20, color='skyblue', edgecolor='black')
  plt.title('Final Grades')
  plt.xlabel('Final Grade)')
  plt.ylabel('Number of Students')
  plt.grid(axis='y', alpha=0.75)
  plt.show()
  plt.figure(figsize=(10, 6))
  sns.scatterplot(data=df, x='studytime', y='G3', color='orange')
  plt.title(' Study Time vs Final Grade')
  plt.xlabel('Study Time ')
  plt.ylabel('Final Grade')
  plt.grid()
  plt.show()
  average_grade_by_gender = df.groupby('sex')['G3'].mean().reset_index()
  plt.figure(figsize=(8, 5))
  sns.barplot(data=average_grade_by_gender, x='sex', y='G3', palette='pastel')
  plt.title('Average  by Gender')
  plt.xlabel('Gender')
  plt.ylabel('Average Final Grade')
  plt.grid(axis='y', alpha=0.75)
  plt.show()




```
